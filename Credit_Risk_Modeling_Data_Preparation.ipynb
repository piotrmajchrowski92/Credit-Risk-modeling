# Data Preparation

### Importing Libraries

```python
import numpy as np
import pandas as pd
```

### Importing Data

```python
loan_data_backup = pd.read_csv('loan_data_2007_2014.csv') # storing the copy of the data before making any changes
```

```python
loan_data = loan_data_backup.copy()
```

### Exploring Data

```python
loan_data
```

```python
pd.options.display.max_columns = None # to see all the columns
```

```python
loan_data
```

```python
loan_data.info()

# term and emp_length are objects, should be numeric
# earlies_cr_line and issue_d should be dates not strings
```

```python
loan_data['emp_length'].unique()
```

```python
loan_data['emp_length_int'].unique()
```

```python
loan_data['emp_length_int'] = loan_data['emp_length'].str.replace('+ years', '')
loan_data['emp_length_int'] = loan_data['emp_length_int'].str.replace('< 1 year', str(0))
loan_data['emp_length_int'] = loan_data['emp_length_int'].str.replace('n/a', str(0))
loan_data['emp_length_int'] = loan_data['emp_length_int'].str.replace(' years', '')
loan_data['emp_length_int'] = loan_data['emp_length_int'].str.replace(' year', '')
```

```python
type(loan_data['emp_length_int'][0])
```

```python
loan_data['emp_length_int'] = pd.to_numeric(loan_data['emp_length_int'])
```

```python
type(loan_data['emp_length_int'][0])
```

```python
loan_data['term'].unique()
```

```python
loan_data['term_int'] = loan_data['term'].str.replace(' months', '') 
```

```python
loan_data['term_int'].unique()
```

```python
loan_data['term_int'] = pd.to_numeric(loan_data['term_int'])
```

```python
loan_data['earliest_cr_line']
```

```python
loan_data['earliest_cr_line_date'] = pd.to_datetime(loan_data['earliest_cr_line'], format = '%b-%y')
```

```python
type(loan_data['earliest_cr_line_date'][0])
```

```python
loan_data['earliest_cr_line_date']
```

```python
pd.to_datetime('2023-12-01') - loan_data['earliest_cr_line_date'] #days since earliest cr line
```

```python
loan_data['mths_since_earliest_cr_line'] = round(pd.to_numeric((pd.to_datetime('2023-12-01') - loan_data['earliest_cr_line_date']) / np.timedelta64(1, 'M')))
```

```python
loan_data['mths_since_earliest_cr_line'].describe()
# Our min value is negative number, so we have a data issue, most likely because the origin of the built_in time scale starts after 1970
```

```python
loan_data.loc[:, ['earliest_cr_line', 'earliest_cr_line_date', 'mths_since_earliest_cr_line']][loan_data['mths_since_earliest_cr_line'] < 0]
```

```python
# to fix this I decided to assign the maximum value instead of negative values
loan_data['mths_since_earliest_cr_line'][loan_data['mths_since_earliest_cr_line'] < 0] = loan_data['mths_since_earliest_cr_line'].max()
```

```python
# now checking the min value
min(loan_data['mths_since_earliest_cr_line'])
```

```python
loan_data['issue_d']
```

```python
loan_data['issue_d_date'] = pd.to_datetime(loan_data['issue_d'], format = '%b-%y')
```

```python
loan_data['issue_d_date']
```

```python
pd.to_datetime('2023-12-01') - loan_data['issue_d_date']
```

```python
loan_data['mths_since_earliest_issue_d_date'] = round(pd.to_numeric((pd.to_datetime('2023-12-01') - loan_data['issue_d_date']) / np.timedelta64(1, 'M')))
```
