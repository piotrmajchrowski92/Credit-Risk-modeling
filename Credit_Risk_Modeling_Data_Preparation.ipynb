### Preprocessing Discrete Variables: Automating Calculations

```python
df_inputs_prepr = loan_data_inputs_train
df_targets_prepr = loan_data_targets_train # dependent variable
```

```python
def woe_discrete(df, discrete_variable_name, good_bad_variable_df):
    df = pd.concat([df[discrete_variable_name], good_bad_variable_df], axis = 1) # this statement will overwrite the original df variable with a two column df - the independent variable and the dependent variable
    df = pd.concat([df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].count(),
                   df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].mean()], axis = 1)
    df = df.iloc[:, [0,1,3]] #keep only specific columns
    df.columns = [df.columns.values[0], 'n_obs', 'prop_good'] #changing column names 
    df['prop_n_obs'] = df['n_obs'] / df['n_obs'].sum() #calculation the proportion of observations for the total number of observations
    df['n_good'] = df['prop_good'] * df['n_obs'] #calculating the number of good borrowers
    df['n_bad'] = (1 - df['prop_good']) * df['n_obs'] #calculationg the number of bad borrowers
    df['prop_n_good'] = df['n_good'] / df['n_good'].sum() #proportion of good borrowers
    df['prop_n_bad'] = df['n_bad'] / df['n_bad'].sum() #proportion of bad borrowers
    df['WoE'] = np.log(df['prop_n_good'] / df['prop_n_bad']) #calculationg weight of evidence
    df = df.sort_values(['WoE'])
    df = df.reset_index(drop = True)
    df['diff_prop_good'] = df['prop_good'].diff().abs()
    df['diff_WoE'] = df['WoE'].diff().abs()
    df['IV'] = (df['prop_n_good'] - df['prop_n_bad']) * df['WoE']
    df['IV'] = df['IV'].sum() #calculating information value
    return df
    
    
```

```python
df_temp = woe_discrete(df_inputs_prepr, 'grade', df_targets_prepr) #first argument is the df where the data is contained, second and third argument are the independent and dependent variables I am interested in
df_temp
```
